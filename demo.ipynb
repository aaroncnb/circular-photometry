```python
import haperflux as hf
import matplotlib.pyplot as plt



```

```python
fd_all, fd_err_all, fd_bg_all = hf.haperfluxMany("regions.list","maps.list",60,100,120)
```

```python
fd_all, fd_err_all_mad, fd_bg_all = hf.haperfluxMany("regions.list","maps.list",60,100,120, noise_model=1)
```

```python
fd_err_ratio = fd_err_all_mad / fd_err_all
```

```python
rel_noise = fd_err_all / fd_all
```

```python
ncols=6
nrows=3
aspect=1.0
import matplotlib.gridspec as gridspec

# fig, axs = plt.subplots(ncols=ncols, 
#                         nrows=nrows, 
#                         sharey=True, 
#                         sharex=True)
# #fig.subplots_adjust(hspace=0.1, left=0.1, right=0.7)

fig = plt.figure(figsize=(11,8))
gs1 = gridspec.GridSpec(nrows, ncols, width_ratios=[1,1,1,1,1,1])

#plt.setp(gs1, aspect=1.0, adjustable='box-forced')

labels = ['A9', 'D12', 'I12', 'A18', 'D25', 'I25', 'D60', 'I60', 'A65',
       'A90', 'D100', 'I100', 'D140', 'A140', 'A160', 'D240', 'P857',
       'P545']

for i in range(0,(ncols*nrows)):
            



    ax = fig.add_subplot(gs1[i])
    #ax.set_aspect(aspect, adjustable='box')
    ax.hist(fd_err_ratio[:,i])
    
    ax.text(0.5, 0.8,labels[i], horizontalalignment='center',
      verticalalignment='center',
      transform=ax.transAxes, 
      fontsize=15)

plt.show()
plt.close()
```
